```python
import pandas as pd
import time
```

```python
start_time = time.time()  # Início da contagem do tempo

df = pd.read_parquet('data/baseDocumentos')

end_time = time.time()  # Fim da contagem do tempo
elapsed_time = end_time - start_time
print(f"Tempo para ler o arquivo: {elapsed_time:.2f} segundos")
```

```python
import json

# Seleciona a primeira linha do DataFrame
instance = df.iloc[27]

# Converte para dicionário e depois para JSON
instance_json = json.dumps(instance.to_dict(), ensure_ascii=False, default=str, indent=2)
print(instance_json)

with open('instance.json', 'w', encoding='utf-8') as f:
    f.write(instance_json)
```

```python
# Mostra onde há valores vazios (NaN) no DataFrame
empty_values = df.isnull().sum()
print(empty_values[empty_values > 0])
print(df.shape)
```
